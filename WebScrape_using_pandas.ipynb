## **Simple Web Scraping on HTML page using pandas**

```python
import pandas as pd

scrape = pd.read_html("https://en.wikipedia.org/wiki/Union_Government_ministries_of_India")

for idx, table in enumerate(scrape):
  print("*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*+*")
  print(idx)
  print(table)
```

```python
# To save this file in CSV_Format
df = pd.DataFrame(scrape[0])
new_df = df.fillna("NaN")
new_df.to_csv("/content/drive/MyDrive/123/DataSet.csv", index=True) # Link would be your google drive path
```
